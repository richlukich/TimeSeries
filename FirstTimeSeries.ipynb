```python
def plot_true_and_another(true_series,another_series,title):
  plt.figure(figsize=(17, 8))
  plt.grid(True)
  plt.plot(true_series,color='b',label='True prices')
  plt.plot(another_series,color='r',label=title)
  plt.legend(loc="best")
  plt.title('Closing price of New Germany Fund Inc (GF)')
  plt.ylabel('Closing price ($)')
  plt.xlabel('Trading day')
  plt.legend()
```

**Скользящая средняя** </br>
$\stackrel{*}{y_t}=\frac{1}{k} \sum_{n=0}^{k-1}y_{t-n} $

```python
def plot_rolling_window(series,n):
  rolling_mean = series.rolling(window=n).mean()

  plot_true_and_another(series,rolling_mean,'Rolling window')
```

Временное окно равно 7 дням

К сожалению сделать долгосрочный прогноз с помощью скользящей средней не удастся ведь мы по сути выражаем текущее значение переменное через предыдущие и текущее значение тоже. Зато с помощью скользящей средней мы можем сгладить исходный временной ряд для выявления трендов. Чем больше зададим ширину интервала — тем более сглаженным окажется тренд. В случае, если данные сильно зашумлены, что особенно часто встречается, например, в финансовых показателях, такая процедура может помочь с определением общих паттернов.

**Взвешенная скользящая средняя** </br>
$\stackrel{*}{y_t}=\sum_{n=1}^{k}w_n\cdot y_{t+1-n} $ </br>
$\sum w_n =1$

```python
def plot_weighted_average(series,weights):
  results=list(series[:len(weights)])
  t=len(weights)
  for i in range (t,len(series)):
    result=0
    for n in range (1,len(weights)+1):
      result+= weights[n-1]*series[i+1-n]
    results.append(result)
  plot_true_and_another(series,results,'Weighted average')
```

```python
weights=[0.5,0.3,0.1,0.05,0.03,0.02]
plot_weighted_average (data.CLOSE,weights)
```

**Экспоненциальное сглаживание, модель Хольта-Винтерса**

**Простое экспоненциальное сглаживание**

$\stackrel{*}{y_t}=\alpha \cdot y_t + (1-\alpha)\cdot \stackrel{*}{y_{t-1}} $

```python
def exponential_smoothing(series, alpha):
  results=[series[0]]
  for i in range (1,len(series)):
    result=alpha*series[i]+(1-alpha)*results[i-1]
    results.append(result)
  return results
  
```

```python
alphas=[0.1,0.5,0.9]
plt.figure(figsize=(17, 8))
plt.plot(data.CLOSE,color='b',label='True prices')
plt.grid(True)
for alpha in alphas:
  plt.plot(exponential_smoothing(data.CLOSE, alpha),label=f'Exponential smoothing Alpha={alpha}')
plt.legend(loc="best")
plt.title('Closing price of New Germany Fund Inc (GF)')
plt.ylabel('Closing price ($)')
plt.xlabel('Trading day')
plt.legend()
```

**Двойное экспоненциальное сглаживание** </br> </br>
$l_x=\alpha y_x+(1-\alpha)(l_{x-1}+b_{x-1}) $ </br>
$b_x=\beta(l_x-l_{x-1})+(1-\beta)b_{x-1} $ </br>
$\stackrel{*}{y_{x+1}}=l_x+b_x $

```python
def double_exponential_smoothing(series, alpha,beta):
  results=[series[0]]
  l=[series[0]]
  b=[series[1]-series[0]]
  for x in range (1,len(series)+1):
    l.append(alpha*series[x-1]+(1-alpha)*(l[x-1]+b[x-1]))
    b.append(beta*(l[x]-l[x-1])+(1-beta)*b[x-1])
    results.append(l[x]+b[x])
  return results    
```

```python
alphas=[0.1,0.9,0.5,0.01]
betas=[0.9,0.1,0.5,0.02]
plt.figure(figsize=(17, 8))
plt.plot(data.CLOSE,color='b',label='True prices')
plt.grid(True)
for alpha,beta in zip(alphas,betas):
  plt.plot(double_exponential_smoothing(data.CLOSE, alpha,beta),label=f'Alpha={alpha}, Beta={beta}')
plt.legend(loc="best")
plt.title('Double Exponential Smoothing')
plt.ylabel('Closing price ($)')
plt.xlabel('Trading day')
plt.legend()
```

Для начала надо проверить наш временной ряд на стационарность, cделаем мы это с помощью теста Дики-Фуллера </br>

Ряд называется стационарным если он представлен  в таком виде: </br>
$y_t= \phi\cdot y_{t-1}+\varepsilon_t,\ $ где $ |\phi|<1 ,\ $ $ \varepsilon_t \sim N(0,\sigma^2) $ 

**Тест Дики Фуллера** </br>


```python
test = sm.tsa.adfuller(data.CLOSE)
print ('adf: ' ,test[0])
print ('p-value: ', test[1])
print ('Critical values: ', test[4])
if test[0]> test[4]['5%']: 
    print ('есть единичные корни, ряд не стационарен')
else:
    print ('единичных корней нет, ряд стационарен')
```

Как неудивительно ряд получился не стационарным. </br>
Построим автокорреляционную функцию

```python
lags=30
figsize=(12, 7)
fig = plt.figure(figsize=figsize)
layout = (2,2)
ts_ax = plt.subplot2grid(layout, (0,0), colspan=2)
acf_ax = plt.subplot2grid(layout, (1,0))
pacf_ax = plt.subplot2grid(layout, (1,1))
        
data.CLOSE.plot(ax=ts_ax)
p_value = sm.tsa.stattools.adfuller(data.CLOSE)[1]
ts_ax.set_title('Time Series Analysis Plots\n Dickey-Fuller: p={0:.5f}'.format(p_value))
sm.tsa.graphics.plot_acf(data.CLOSE, lags=lags, ax=acf_ax)
sm.tsa.graphics.plot_pacf(data.CLOSE, lags=lags, ax=pacf_ax)
plt.tight_layout()
```

Попробуем убрать стационарность с помощью преобразования Бокса-Кокса

```python
def invboxcox(y,lmbda):
    if lmbda == 0:
        return(np.exp(y))
    else:
        return(np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
plt.figure(figsize=(17, 8))
plt.grid(True)
plt.plot(invboxcox(data.CLOSE,lmbda = 1.6),color='b')
plt.title('Box-Cox Transform')
plt.ylabel('Closing price ($)')
plt.xlabel('Trading day')
```

Как видно не получается не получается сделать ряд стационарным с помощью преобразования Бокса-Кокса. Попробуем тогда вычесть из исходного ряда сам исходный ряд с задержкой один день

```python
data_diff = data.CLOSE - data.CLOSE.shift(1)
```

```python
lags=30
figsize=(12, 7)
fig = plt.figure(figsize=figsize)
layout = (2,2)
ts_ax = plt.subplot2grid(layout, (0,0), colspan=2)
acf_ax = plt.subplot2grid(layout, (1,0))
pacf_ax = plt.subplot2grid(layout, (1,1))
        
data_diff[1:].plot(ax=ts_ax)
p_value = sm.tsa.stattools.adfuller(data_diff[1:])[1]
ts_ax.set_title('Time Series Analysis Plots\n Dickey-Fuller: p={0:.5f}'.format(p_value))
sm.tsa.graphics.plot_acf(data_diff[1:], lags=lags, ax=acf_ax)
sm.tsa.graphics.plot_pacf(data_diff[1:], lags=lags, ax=pacf_ax)
plt.tight_layout()
```

Мы добились стационарности и можем перейти к моделированию!
